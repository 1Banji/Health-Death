```python
from health_death_lib import Eda
from health_death_lib import Prep
from health_death_lib import Model_Selection
from health_death_lib import Test
from sklearn.linear_model import LinearRegression
import pandas as pd
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings(action='ignore')
```

```python
xcl = pd.ExcelFile('2019_County_Health_Rankings_Data_v2.xls')
```

```python
#Get dataframes from excel by sheetnames
ranked_measure = Eda.get_df('Ranked Measure Data',xcl)
addit_measure = Eda.get_df('Additional Measure Data',xcl)

#Get dataframe by neccessary columns
rank_df = Eda.get_ranked_df(ranked_measure)
add_df = Eda.get_add_df(addit_measure)

#Merge and drop the null values
df_merged = Eda.merge_dropna(add_df,rank_df)

#Get numerical columns
num_df = Eda.get_num_df(df_merged)
```

## Preprocessing

```python
train_set, test_set = train_test_split(num_df, test_size=0.4, random_state=42)
```

```python
#Remove the outliers
train_set = Prep.remove_outliers(train_set)
```

```python
df_seven = Eda.get_df_seven(train_set)
```

There is a high chance of multicollinearity between two independent variables % Smokers, %Frequent Mental Distress and % Frequent Physical Distress and % Fair/Poor as we see from the heatmap above they are highly related to each other.

```python
Prep.normality_check(df_seven)
```

Data doesn't look normal but for Linear Regression we don't need to normalize it since we will use Standard Scale.

```python
target = train_set['Life Expectancy']
pred_variables = train_set[['% Smokers', 'Teen Birth Rate', '% Diabetic','% Physically Inactive', '% Fair/Poor' ]]
```

```python
pred_df_seven = Prep.scale(pred_variables)
```

We will run different regression models to see which one performs best for our task and choose a baseline model among them

## Model Selection

```python
lin_rmse, lin_r2 = Model_Selection.lin_model(pred_df_seven,target)
Model_Selection.display_scores('Linear Model',lin_rmse, lin_r2)
```

This is our baseline model. We will try to improve the model by adding more attributes. We will use only the attributes which has more than 0.4 correlation with Life Expectancy to keep our model as simple as possible. Before that we will check training data to make sure if there is any multicollinearity between independent variable.

```python
tree_rmse, tree_r2 = Model_Selection.decision_tree_model(pred_df_seven, target)
Model_Selection.display_scores('Decision Tree Model', tree_rmse, tree_r2)
```

```python
forest_rmse, forest_r2 = Model_Selection.random_forest_model(pred_df_seven,target)
Model_Selection.display_scores('Random Forest', forest_rmse, forest_r2)
```

```python
alpha_values = [.05, .1, .15, .2, .3 , .4]
for i in alpha_values:
    lasso_rmse, lasso_r2 = Model_Selection.lasso_model(pred_df_seven, target, i)
    name = 'Lasso ' + str(i)
    Model_Selection.display_scores(name,lasso_rmse, lasso_r2)
    print('\n')
```

Our baseline model will be the Linear Regression model.  We will try to improve the model by adding more attributes. We will use only the attributes which has more than 0.4 correlation with Life Expectancy to keep our model as simple as possible. Before that we will check training data to make sure if there is any multicollinearity between independent variable.

## Preprocessing before improving the model

```python
corr, df_cor = Prep.correlated(train_set)
```

```python
Prep.multicollinear(corr)
```

```python
multicollinear = ['% Frequent Physical Distress', '% Frequent Mental Distress']
predictors = df_cor.drop(multicollinear, axis=1)
predictors.drop('Life Expectancy', axis=1, inplace=True)
```

```python
Prep.scatter_plots(predictors,target)
```

```python
predictors.drop(['% Screened', 'Food Environment Index', 'Average Daily PM2.5'],axis=1, inplace=True)
```

## Running a new model

```python
pred_scaled = Prep.scale(predictors)
```

```python
lin2_rmse, lin2_r2 = Model_Selection.lin_model(pred_scaled, target)
Model_Selection.display_scores("Linear Model 2", lin2_rmse, lin2_r2)
```

We improved our model comparing with our baseline model

```python
ols_result = Model_Selection.ols(pred_scaled, target)
ols_result.summary()
```

```python
Model_Selection.ols_residual_check(ols_result, target, 0.05)
```

## Adding polynomial features

```python
degrees = [1,2,3,4]
for degree in degrees:
    poly_rmse, poly_r2 = Model_Selection.polynomial(degree,pred_scaled, target)
    name = 'Poly ' + str(degree)
    Model_Selection.display_scores(name, poly_rmse, poly_r2)
    print('\n')
```

Adding polynomial features also didn't improve our model as well.

```python
x_test, y_test = Test.test_data(test_set, predictors)
lin_final = LinearRegression().fit(pred_scaled,target)
x_test = Prep.scale(x_test)
predicted_final = lin_final.predict(x_test)
```

```python
Test.residual_plot(lin_final, pred_scaled, target, x_test, y_test)
```
